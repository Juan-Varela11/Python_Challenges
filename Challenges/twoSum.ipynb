# The Problem

Find the number of different combinations of two numbers in some list which sum up to some specified number.

# The Solution 

```python
import random
import matplotlib.pyplot as plt
```

```python
# Counting number of different combinations which add up to some number

number_list = [2,1,3,4,5,2,2,4,67,67,6,34,23,23,43,2,1,3,435,4,3] # list of random numbers 
count = [] # keep track of amount of combinations
goal = 6 # goal number 

def twoSum_1(array1):
    for i in range(len(array1)):
        for j in range(len(array1)):
            check = array1[i] + array1[j]
            if check == goal:
                print('{} plus {} equals {}.'.format(array1[i],array1[j],array1[i]+array1[j]))
                count.append(1)
    print('\nThere are {} combinations that sum up to {} from the given numbers list.'.format(sum(count),goal))
        
twoSum_1(number_list)
```

```python
# Counting and visualzing the number of combinations for different goals

number_list = list(range(0,1001)) # list of numbers from 0 to 1000, inclusive
count = []; goal_list = []; count_list = [];

def twoSum_2(array2):
    for k in range (0,50): # number of  different goals being tested
        goal = random.randrange(0,1001) # goal can take a value from 0 to 1000
        for i in range(len(array2)):
            for j in range(len(array2)):
                check = array2[i] + array2[j]
                if check == goal:
                    count.append(1)
        count_list.append(sum(count)) #factor of 0.5 would need to be included to account for repeats                      
        goal_list.append(goal)                                                                      
        count.clear()
        
twoSum_2(number_list)

# Visualizing the number of combinations to reach different goals
fig = plt.figure()
plt.scatter(goal_list, count_list)
plt.title('Amount of Two Number Combinations to Sum to a Goal')
plt.xlabel('Numeric Goal')
plt.ylabel('Combinations')
plt.show()
```
